```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Feature engineering

We will do the following feature engineering exercises to extract meaningful information from data:
<ul>
<li>Creating an aggregate view of Trips<br></li>
<li>Calculating trip times in seconds, minutes, hours and even days to find outliers<br></li>
<li>Calculating distances between stations using latitudes and longitudes<br></li>
<li>Calculating station utilization using the following formula: <br>[# of docks available] / [# of bikes available + # of docks available]<br></li>
</ul>

##### Saved view 1 -- aggregating trip details and adding additional columns such as distance between stations (using Google ST_GeogPoint()) and trip times 

Saved view: `profound-surf-264703.bike_trip_data.bike_trip_distance_duration_station_view`

```python
%%bigquery df_trip_details

        select 
            a.trip_id, 
            round(a.duration_sec, 2) as duration_sec, 
          round(a.duration_sec/60, 3) as duration_min, 
          round(a.duration_sec/(60*60), 3) as duration_hour, 
          round(a.duration_sec/(60*60*24), 2) as duration_day, 
            a.start_date, 
            a.start_station_name, 
            a.start_station_id, 
            a.end_date, 
            a.end_station_name, 
            a.end_station_id, 
            abs(a.start_station_id - a.end_station_id),
            a.zip_code, 
            a.subscriber_type,
            EXTRACT(DAYOFWEEK FROM a.start_date) AS dow_int,
            CASE EXTRACT(DAYOFWEEK FROM a.start_date)
                   WHEN 1 THEN "Sunday"
                   WHEN 2 THEN "Monday"
                   WHEN 3 THEN "Tuesday"
                   WHEN 4 THEN "Wednesday"
                   WHEN 5 THEN "Thursday"
                   WHEN 6 THEN "Friday"
                   WHEN 7 THEN "Saturday"
                   END AS dow_str,
                   CASE 
                   WHEN EXTRACT(DAYOFWEEK FROM a.start_date) IN (1, 7) THEN "Weekend"
                   ELSE "Weekday"
                   END AS dow_weekday,
                   EXTRACT(HOUR FROM a.start_date) AS start_hour,
                   CASE 
                   WHEN EXTRACT(HOUR FROM a.start_date) <= 5  OR EXTRACT(HOUR FROM a.start_date) >= 23 THEN "Nightime"
                   WHEN EXTRACT(HOUR FROM a.start_date) >= 6 and EXTRACT(HOUR FROM a.start_date) <= 8 THEN "Morning"
                   WHEN EXTRACT(HOUR FROM a.start_date) >= 9 and EXTRACT(HOUR FROM a.start_date) <= 10 THEN "Mid Morning"
                   WHEN EXTRACT(HOUR FROM a.start_date) >= 11 and EXTRACT(HOUR FROM a.start_date) <= 13 THEN "Mid Day"
                   WHEN EXTRACT(HOUR FROM a.start_date) >= 14 and EXTRACT(HOUR FROM a.start_date) <= 16 THEN "Early Afternoon"
                   WHEN EXTRACT(HOUR FROM a.start_date) >= 17 and EXTRACT(HOUR FROM a.start_date) <= 19 THEN "Afternoon"
                   WHEN EXTRACT(HOUR FROM a.start_date) >= 20 and EXTRACT(HOUR FROM a.start_date) <= 22 THEN "Evening"
                   END AS start_hour_str,
            b.latitude as start_station_lat, 
            b.longitude as start_station_long , 
            b.dockcount as start_station_dockcount, 
            b.landmark as start_station_landmark, 
            b.installation_date as start_station_install_date, 
            c.latitude as end_station_lat, 
            c.longitude as end_station_long, 
            ROUND(ST_Distance(ST_GeogPoint(b.longitude, b.latitude), ST_GeogPoint(c.longitude, c.latitude)) * 0.00062137, 2) as distance_between_stations

        FROM 
            `bigquery-public-data.san_francisco.bikeshare_trips` a, 
            `bigquery-public-data.san_francisco.bikeshare_stations` b, 
            `bigquery-public-data.san_francisco.bikeshare_stations` c
        where 
            a.start_station_id = b.station_id and 
            a.end_station_id = c.station_id
        ORDER BY 
            distance_between_stations desc, duration_sec desc
```

##### Saved view 2 -- calculate station utilization rates grouped by by time-slots of day (e.g. nighttime, morning, evening...), date, station_id and other metrics 

Saved view: `profound-surf-264703.bike_trip_data.station_utilization_by_date_time`

```python
%%bigquery df_station_utilization

select a.station_id, 
       round(((docks_available)/ NULLIF((bikes_available + docks_available) , 0)),2) as station_util, 
       b.latitude,
       b.longitude,
        CASE 
               WHEN EXTRACT(HOUR FROM a.time) <= 5  OR EXTRACT(HOUR FROM a.time) >= 23 THEN "Nightime"
               WHEN EXTRACT(HOUR FROM a.time) >= 6 and EXTRACT(HOUR FROM a.time) <= 8 THEN "Morning"
               WHEN EXTRACT(HOUR FROM a.time) >= 9 and EXTRACT(HOUR FROM a.time) <= 10 THEN "Mid Morning"
               WHEN EXTRACT(HOUR FROM a.time) >= 11 and EXTRACT(HOUR FROM a.time) <= 13 THEN "Mid Day"
               WHEN EXTRACT(HOUR FROM a.time) >= 14 and EXTRACT(HOUR FROM a.time) <= 16 THEN "Early Afternoon"
               WHEN EXTRACT(HOUR FROM a.time) >= 17 and EXTRACT(HOUR FROM a.time) <= 19 THEN "Afternoon"
               WHEN EXTRACT(HOUR FROM a.time) >= 20 and EXTRACT(HOUR FROM a.time) <= 22 THEN "Evening"
               END 
        AS start_hour_str,
      EXTRACT(DATE FROM a.time) as util_date
from `bigquery-public-data.san_francisco.bikeshare_status` a,
      `bigquery-public-data.san_francisco.bikeshare_stations` b
where a.station_id = b.station_id
group by station_id, station_util, start_hour_str, latitude, longitude, util_date
order by station_util DESC
```

#### 1- Commuter trips

Commuter trips can be defined as trips with the following characteristics: 
<ul>
<li>Trips originating during the rush hour times (early mornings, mornings, afternoons, late afternoons and late evenings)<br></li>
<li>Typically short distances (i.e. less than 5 miles)<br></li>
<li>Only originating and ending on weekdays<br></li>
<li>Start station not being same as the end station i.e. distance between stations is not zero (to eliminate leisure trips from tourists or returned bikes)<br></li>
</ul>

We could possibly add another criteria of filtering out only paid "annual subscribers". However, applying that criteria might lead to an inadvertent removal of commuters that haven't yet renewed their annual memberships or those who or sparringly use bikes for commute to workplaces. Hence for our purposes, we are not applying that criteria. 

Some data scientists might want to also combine "trip pairs" i.e. combine the combination of start and end stations and using the pair as one "trip". But that might not be a good idea considering that the bike rentals are "one-way rentals"; and commuters might only use bikes for one section of their office journey (and could use public transit or shared cabs for the other leg of their journey). Including both legs of the to-n-fro journey as separate would therefore give us a more holistic view of the most commonly used station pairs during rush hours. 


```python
%%bigquery df

        select   
            start_station_name,
            end_station_name,
            distance_between_stations as distance_bet_stations_miles,
            round(avg(duration_min),2) as avg_trip_duration_minutes,
            round((count(*) / (SELECT count(*) FROM 
                                   `profound-surf-264703.bike_trip_data.bike_trip_distance_duration_station_view`
                                   where 
                                    start_station_name <> end_station_name
                                      and distance_between_stations <> 0
                                      and start_hour_str in ('Mid Morning', 'Morning', 'Evening', 'Afternoon', 'Early Afternoon')
                                      and dow_weekday='Weekday')),5) AS Pct_To_Total
        FROM 
            `profound-surf-264703.bike_trip_data.bike_trip_distance_duration_station_view` a
        where 
              start_station_name <> end_station_name
              and distance_between_stations <> 0
              and start_hour_str in ('Mid Morning', 'Morning', 'Evening', 'Afternoon', 'Early Afternoon')
              and dow_weekday='Weekday'
        group by distance_between_stations, start_station_name, end_station_name
        ORDER BY Pct_To_Total desc 
```

```python
df
```

We see that even with applying our "commute trip" criteria, there are a lot of outlier records that skew our dataset. For example, the below histograms show that there are few trips that originate on a weekday with a travel distance greater than 40 miles! Clearly, these are only very special cases which can be excluded from our analysis. Additionally, we see that almost all commute trips are under 10 miles and less than 200 minutes.

```python
sns.distplot(df['distance_bet_stations_miles'], kde=False, color='red', bins=200)
plt.title('Trip Distances', fontsize=18)
plt.xlabel('Dist between stations in miles', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
```

```python
sns.distplot(df['avg_trip_duration_minutes'], kde=False, color='red', bins=15)
plt.title('Average Trip Durations (minutes)', fontsize=18)
plt.xlabel('Average times between trips', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
```

#### 2 - Station utilization and bottlenecks -- and possibly finding ways to route usage to less utilized stations

Following query utilizes saved views above to find alternate stations around the most heavily used stations within a 0.2 mile radius.   

```python
%%bigquery df_alternate_low_util_stations

select 
  a.station_id as start_stn_id, 
  a.util_date as trip_date, 
  round(avg(a.station_util),3) as start_stn_util, 
  a.start_hour_str as start_stn_hour, 
  a.latitude as start_stn_lat, 
  a.longitude as start_stn_long,
  c.station_id as near_stn_id, 
  round(avg(c.station_util),3) as near_stn_util,    
  c.latitude as near_stn_lat, 
  c.longitude as near_stn_long,
  avg(ROUND(ST_Distance(ST_GeogPoint(a.longitude, a.latitude), ST_GeogPoint(c.longitude, c.latitude)) * 0.00062137, 4)) as near_stn_distance,
  b.trip_id as cust_trip_id, 
  b.zip_code as cust_zipcode, 
  avg(b.distance_between_stations) as cust_trip_distance, 
  b.end_station_id, 
 from 
    `profound-surf-264703.bike_trip_data.station_utilization_by_date_time` a,
     `profound-surf-264703.bike_trip_data.bike_trip_distance_duration_station_view` b,
    `profound-surf-264703.bike_trip_data.station_utilization_by_date_time` c
 where a.station_id = b.start_station_id 
  and a.util_date = EXTRACT(DATE FROM b.start_date )
  and a.start_hour_str = b.start_hour_str 
  and b.distance_between_stations <> 0
  and c.util_date = a.util_date
  and c.station_id <> a.station_id
  and c.start_hour_str = a.start_hour_str
  and c.station_util < a.station_util
  and ROUND(ST_Distance(ST_GeogPoint(a.longitude, a.latitude), ST_GeogPoint(c.longitude, c.latitude)) * 0.00062137, 3) <= 0.2
 group by 
  a.station_id, 
  a.util_date, 
  a.start_hour_str, 
  a.latitude, 
  a.longitude,
  c.station_id, 
  c.latitude, 
  c.longitude,
  b.trip_id, 
  b.zip_code, 
  b.end_station_id
  order by a.util_date desc, start_stn_util desc, near_stn_util asc, near_stn_distance asc, cust_trip_id
```

```python
df_alternate_low_util_stations
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,3))

df_alternate_low_util_stations['start_stn_util'].plot(kind='hist', ax=axs[0], title='Util% of Start Station')
df_alternate_low_util_stations['start_stn_hour'].value_counts().plot(kind='bar', ax=axs[1], title='Peak hour slots')
df_alternate_low_util_stations['near_stn_util'].plot(kind='hist', ax=axs[2], title='Util% of Nearby Station')
```

Looking at the above results, we can easily deduce that the most "in demand" start stations are usually most busy with up to 100% utilization rates around rush hour times such as afternoons, mornings and early afternoons.

We can also see in the 3rd subplot of corresponding stations within a 0.2 mile radius, that most of these stations are only up to 50% full even during peak rush hour times. Certainly something could be done to move demand to these under utilized stations. 

#### 3 - Cancelled commuter trips

Here we are analyzing dataset of trips where start station is equal to end station (i.e. trip distance = 0). The idea behind this exploration would be to understand if there are trips where customers are forced to return their bikes due to an involuntary condition outside of the control of the customer, most likely where the bike is in poor condition and if the bike breaks down.

Hence, to filter out such trips where the customer rents out a bike and the bike breaks down within a few minutes, we will need to apply the following criteria: 
<ul>
<li>Trips originating and ending in the same station<br></li>
<li>Trips originating during rush hour commute (early mornings, mornings, afternoons, late afternoons and late evenings)<br></li>
<li>Trip durations less than 15 minutes<br></li>
<li>Customer is a annual subscriber (not a one time customer -- though customer satisfaction for such one-time customers is also important, but not a focus for this exercise here) <br></li>
</ul>

```python
%%bigquery df_trips_broken_bikes

select 
     distinct a.trip_id, 
     extract(date from a.start_date) as start_date, 
     a.duration_min, 
     a.start_hour_str, 
     a.start_station_name, 
     a.zip_code
    from 
    `profound-surf-264703.bike_trip_data.bike_trip_distance_duration_station_view` a
     where 
     a.distance_between_stations = 0 and 
      a.duration_min <=15 
      and a.subscriber_type = 'Subscriber'
      and a.start_hour_str  in  ('Mid Morning', 'Morning', 'Evening', 'Afternoon', 'Early Afternoon')
     order by start_date desc
```

```python
df_trips_broken_bikes
```

```python
fig, axs = plt.subplots(1,2, figsize=(12,3))

df_trips_broken_bikes['duration_min'].plot(kind='hist', ax=axs[0], title='Trip durations in minutes before return')
df_trips_broken_bikes['start_hour_str'].value_counts().plot(kind='bar', ax=axs[1], title='Return hours of day')
```

With the above data and histograms, we do see that on a daily basis, at least 3 breakdowns occur (for a total of ~4200 returns in 4 years) for paid customers.The most common times of such returns being within 4 minutes of starting the trip. Also, we do see that consistent with our expectations, most of the returns happen during the peak rush hour afternoon slot.

Therefore,  much can be done to improve customer satisfaction during for such circumstances. 

### FINAL RECOMMENDATIONS

#### 1 - Focus most paid rush hour commuter trips

As seen in some of the earlier sections and data pulls (in the README.md file - section 2), 86% of all trips originate from paid subscribers. Out of those, bulk of the trips are weekday commute trips. Hence it is imperative that most the attention is paid on such "paid daily commuter trips". 

As a consequence, the primary recommendation is to solve utilization bottleneck at the busiest bike stations. As seen in the above datapulls, some of the stations can be 100% utilized during peak times, which is bound to impact customer satisfaction and subsequently revenues. 

Also, we see that a bulk of the busiest stations are flanked by correspondingly under-utilized stations with ~50% utilization rates. So perhaps Lyft could create a <B>"Customer Incentive" model</B>. The company could incentivize customers with small dollar gifts or free additional trips for using under-utilized bike stations during peak times (perhaps sending them reminders on their mobile apps a night before).

This would drive the demand to those under utilized stations and create a more equitable use of all bikes and all stations. 

#### 2 - Focus on bike breakdown conditions, bike returns and customer satisfaction

As seen in the last EDA exercise (canceled trips), we do see a consistent volume of 3 to 4 bike breakdowns during rush hour commute. This is bound to create a lot of anger, frustration and customer dissatisfaction. 

The company is recommended to replenish new bikes at the most heavily utilized station routes and make sure that a customer-focused and well-trained staff is present during rush hours to accommodate immediate breakdown scenarios. This will go a long way is generating addition goodwill and revenue.
